#Loop Controle Statements

```python
#Break statement
list_name = "abdulrehmanabbbas"
for listLetter in list_name:
  if listLetter=='a' or listLetter=='n':
      break

print(f'Current letter is :{listLetter}')
```

```python
#Continue Statement
list_name = "abdulrehmanabbbas"
for listLetter in list_name:
  if listLetter=='i' or listLetter=='e':
    continue

print(f'current letter is : {listLetter}')
```

#Classes and Objects

#classes

```python
class Person:
  def _init_(self, name, age):
    self.name=name
    self.age=age
  def myFunction(self):
    print("my name is "+ self.name)



p1 = Person()
p1.name = "abdulrehman "
p1.age = 23
p1.myFunction()

```

```python
class Pharmacist:
  def _init_(self,age,name,degree,passing_year):
    self.name="abdulrehman"
    self.age=23
    self.degree = "None"
    self.passing_year = 0

  def degree_type(self,degree):
    self.degree=degree
    print("degree type is "+ self.degree)


  def passing_year(self, year):
    self.passing_year = year
```

#Objects

```python
pharmacist = Pharmacist()
pharmacist.name = "abdulrehman"
pharmacist.degree = "B_category"
pharmacist.passing_year = 2024
pharmacist.age = 23
print(pharmacist.name)
print(pharmacist.age)
pharmacist.degree_type("c_category")
```
